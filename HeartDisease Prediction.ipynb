```python
from matplotlib.cm import rainbow
from matplotlib import rcParams
```

```python
rcParams['figure.figsize']=10,8
plt.bar(data['target'].unique(),data['target'].value_counts(), color=['red','green'])
plt.xticks([0,1])
plt.ylabel('count')
plt.xlabel('target values')
```

```python

from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
data[scale] = standardScaler.fit_transform(data[scale])
```

```python
data
```

```python
y = data['target']
X = data.drop(['target'], axis = 1)
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
#KNN Classification
from sklearn.neighbors import KNeighborsClassifier
scores = []
for k in range(1,15):
    classifier = KNeighborsClassifier(n_neighbors = k)
    classifier.fit(X_train, y_train)
    scores.append(classifier.score(X_test, y_test))
plt.plot([k for k in range(1, 15)], scores, color = 'blue')
for i in range(1,15):
    plt.text(i, scores[i-1], (i, scores[i-1]))
plt.xticks([i for i in range(1, 21)])
plt.xlabel('No.of Neighbors')
plt.ylabel('Scores')
```

```python
clf = KNeighborsClassifier()
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm_test = confusion_matrix(y_test,y_pred)
cm_test
```

```python
accuracy = clf.score(X_test,y_test)
print(accuracy*100)
```
